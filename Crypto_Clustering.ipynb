```python
# Import required libraries and dependencies
import pandas as pd
import numpy as np
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

# Fit the scaler to the data and transform it
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
scaled_df = pd.DataFrame(
    scaled_data,
    columns=df_market_data.columns,
    index=df_market_data.index
)

# Copy the crypto names from the original DataFrame
if 'coin_id' in df_market_data.columns:
    scaled_df['coin_id'] = df_market_data['coin_id']
  
# Set the coin_id column as index
    scaled_df.set_index('coin_id', inplace=True)
else:
    scaled_df.index = df_market_data.index

# Display the scaled DataFrame
display(scaled_df.head())
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):
    
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    
    # 2. Fit the model to the data using scaled_data
    kmeans.fit(scaled_df)

    # 3. Append the model.inertia_ to the inertia list
    inertia.append(kmeans.inertia_)

# Print the results
print("Inertia values for k=1 to 11:")
for k, i in zip(range(1, 12), inertia):
    print(f"k={k}: Inertia = {i:.2f}")
    
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': list(range(1,12)),
    'inertia': inertia
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Display the DataFrame
display(elbow_df)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve = elbow_df.hvplot.line(
   x="k", y="inertia", label="Raw Data", color="blue", line_dash='solid'
)

# Print the plot
elbow_curve
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 6

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
optimal_k = 6
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
clusters = kmeans.predict(scaled_df)

# Print the resulting array of cluster values.
print(clusters)
```

```python
# Create a copy of the scaled DataFrame
clustered_df = scaled_df.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
clustered_df['Cluster'] = clusters

# Display the copy of the scaled DataFrame
display(clustered_df.head())
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = clustered_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Cluster",  
    hover_cols=["coin_id"],
    title="Cryptocurrency Clusters by 24h vs 7d Price Changes",
    width=800,
    height=500,
    alpha=0.7,
    size=100,   
    legend='top_right',
    colorbar=True,
    cmap='Category10' 
)

# Display the plot
scatter_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(scaled_df)

# View the scaled PCA data
pca_df = pd.DataFrame(
    data=pca_data,
    columns=["PC1", "PC2", "PC3"],
    index=scaled_df.index  
)

# Print the scaled PCA data 
print(pca_df)
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Create variance explanation table
variance_df = pd.DataFrame({
    'Principal Component': ["PC1", "PC2", "PC3"],
    'Individual Variance': explained_variance,
    'Cumulative Variance': cumulative_variance
}).set_index('Principal Component')

# Display results
display(variance_df)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89.50%

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(
    data=pca_data,
    columns=[f"PC{i+1}" for i in range(pca_data.shape[1])]
)

# Copy the crypto names from the original scaled DataFrame
pca_df['coin_id'] = df_market_data.index.values

# Set the coin_id column as index
pca_df.set_index('coin_id', inplace=True)

# Display the scaled PCA DataFrame
display(pca_df.head())
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    
    # Inside the loop:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)

    # 2. Fit the model to the data using `pca_df`
    kmeans.fit(pca_df)

    # 3. Append the model.inertia_ to the inertia list
    inertia.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_optimized = {
    "k": k_values,  
    "inertia": inertia        
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_optimized = pd.DataFrame(elbow_data_optimized)

# Display the DataFrame
display(elbow_data_optimized)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_plot_optimized = elbow_data_optimized.hvplot.line(
    x="k", y="inertia", label="Optimized Data", color="green", line_dash='dashed'
)

# Display the plot
elbow_plot_optimized
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4. This suggests that 4 clusters capture the majority of the data. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Yes, the original data suggested k=6, while PCA reduced it to k=4. 

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans_pca = KMeans(n_clusters=4, random_state=42, n_init=10)  
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
clusters_labels = kmeans_pca.predict(pca_df)

# Print the resulting array of cluster values.
print(clusters_labels)
```

```python
# Create a copy of the scaled PCA DataFrame
pca_df_clusters = pca_df.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
pca_df_clusters['Cluster'] = kmeans_pca.labels_

# Display the copy of the scaled PCA DataFrame
display(pca_df_clusters)
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# The cryptocurrency represented by each data point.
scatter_plot_optimized = pca_df_clusters.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="Cluster",  
    hover_cols=["coind_id"],
    title="Cryptocurrency Clusters by PCA Components",
    width=800,
    height=500,
    alpha=0.7,
    size=100,   
    legend='top_right',
    colorbar=True,
    cmap='Category10' 
)

# Display the plot
scatter_plot_optimized
```
